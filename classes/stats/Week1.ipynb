```r
Mode <- function(x) {
  ux <- unique(x)
  nx <- tabulate(match(x, ux))
  maxid <- which(nx == max(nx))
  if(length(maxid) > 1) {
    warning("There are multiple modes, only the first value is returned, other modes are at: ",paste(ux[maxid[2:length(maxid)]],collapse=", "))
  }
  ux[maxid[1]]
}

c(mean(d$IQ), Mode(d$IQ), median(d$IQ))
```

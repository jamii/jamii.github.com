```r
confidence.interval <- function (estimate, interval, n, sse) {
    error <- sqrt(qf(interval, df1=1, df2=n-1)) * sqrt(sse / (n-1)) / sqrt(n)
    c(estimate-error, estimate+error)
}

confidence.interval(115.14, interval=0.90, n=40, sse=5850)
confidence.interval(115.14, interval=0.95, n=40, sse=5850)
confidence.interval(115.14, interval=0.95, n=79, sse=5850)
confidence.interval(105.14, interval=0.95, n=40, sse=5850)
```

```r
library(foreign)

d <- as.data.frame(read.spss("./IQscores.sav"))
summary(d)
```

```r
pre <- function (actual, estimate.c, estimate.a) {
    sse.c <- sum((estimate.c - actual) ** 2)
    sse.a <- sum((estimate.a - actual) ** 2)
    (sse.c - sse.a) / sse.c
}

pre(d$IQ, 100, mean(d$IQ))

f.score <- function (actual, estimate.c, estimate.a, pc, pa) {
    p = pre(actual, estimate.c, estimate.a) 
    (p / (pa - pc)) / ((1 - p) / (length(actual) - pa))
}

f.score(d$IQ, 100, mean(d$IQ), 0, 1)

significance <- function(actual, estimate.c, estimate.a, pc, pa) {
    1 - pf(f.score(actual, estimate.c, estimate.a, pc, pa), df1=1, df2=length(actual)-1)
}

significance(d$IQ, 100, mean(d$IQ), 0, 1)
```

```r
Mode <- function(x) {
  ux <- unique(x)
  nx <- tabulate(match(x, ux))
  maxid <- which(nx == max(nx))
  if(length(maxid) > 1) {
    warning("There are multiple modes, only the first value is returned, other modes are at: ",paste(ux[maxid[2:length(maxid)]],collapse=", "))
  }
  ux[maxid[1]]
}

guesses <- d[!is.na(d$Lecturer_height),]$Lecturer_height
pre(196, guesses, guesses - guesses + mean(guesses))
pre(196, guesses, guesses - guesses + Mode(guesses))
pre(196, guesses, guesses - guesses + median(guesses))
c(mean(guesses), Mode(guesses), median(guesses))

guesses <- d[!is.na(d$Lecturer_age),]$Lecturer_age
pre(42, guesses, guesses - guesses + mean(guesses))
pre(42, guesses, guesses - guesses + Mode(guesses))
pre(42, guesses, guesses - guesses + median(guesses))
c(mean(guesses), Mode(guesses), median(guesses))
```

```r
library(foreign)
d <- as.data.frame(read.spss("./2017_Questionnaire_1.sav"))
d <- d[,c("Weight", "Height","Age")]
d <- d[complete.cases(d),]
summary(d)
pairs(d)
```

```r
model <- lm(Weight ~ Height + Age, data=d)
summary(model)
```

```r
predictors <- as.matrix(d[,c("Height","Age")])
mahal <- mahalanobis(predictors, colMeans(predictors), cov(predictors))
d[mahal >= qchisq(1-.05,df=ncol(predictors)),]
```

```r
lever <- hatvalues(model)
d[lever > (2*4)/nrow(d),]
```

```r
sdr <- rstudent(model)
qt(1-(.05/2)/nrow(d),df=nrow(d)-4-1)
d[abs(sdr) >= qt(1-(.05/2)/nrow(d),df=nrow(d)-4-1),]
```

```r
cooksd <- cooks.distance(model)
summary(cooksd)
d[cooksd >= 2,]
```

```r
d2 <- d[!(mahal >= qchisq(1-.05,df=ncol(predictors))),]
nrow(d)
model2 <- lm(Weight ~ Height + Age, data=d2)
summary(model2)
```

```r
rs <- residuals(model2)
mean(rs) / sd(rs)
hist(rs,prob=TRUE, breaks=20)
curve(dnorm(x,mean=0,sd=sd(rs)),add=TRUE)
```

```r
d <- as.data.frame(read.spss("./reactionIQ.sav"))
d <- d[complete.cases(d),]
summary(d)
pairs(d)
```

```r
model <- lm(RT ~ AH4, data=d)
summary(model)
sse <- sum(residuals(model)^2)
```

```r
d$AH4.2 <- d$AH4 ^ 2
model2 <- lm(RT ~ AH4 + AH4.2, data=d)
summary(model2)
sse2 <- sum(residuals(model2)^2)
pre <- (sse - sse2) / sse
pre
```

```r
anova(model2, model)
```

```r
sdr <- rstudent(model2)
threshold <- qt(1-(.05/2)/nrow(d),df=nrow(d)-3-1)
d[abs(sdr) >= threshold,]
d <- d[!(abs(sdr) >= threshold),]
```

```r
model3 <- lm(RT ~ AH4 + AH4.2, data=d)
summary(model3)
plot(model3)
```

```r
hist(residuals(model3), prob=TRUE, breaks=30)
curve(dnorm(x,mean=0,sd=sd(residuals(model3))),add=TRUE)
```

```r
rs <- residuals(model3)
shapiro.test(rs)
ks.test(rs, "pnorm", mean=0, sd=sd(rs))
```

```r
null <- lm(RT ~ 1, data=d)
sse.null <- sum(residuals(null)^2)
sse.3 <- sum(residuals(model3)^2)
ssr.y <- sse.null - sse.3
g <- residuals(model3)^2 / (ssr.y / nrow(d))
hist(g)
d$g <- g
bp <- lm(g ~ AH4 + AH4.2, data=d)
sse.bp <- sum(residuals(bp)^2)
bp.null <- lm(g ~ 1)
summary(bp.null)
sse.bp.null <- sum(residuals(bp.null)^2)
ssr.g <- sse.bp.null - sse.bp
1-pchisq(ssr.g/2,df=2)
```

```r
d$log.RT <- log(d$RT)
pairs(d)
```

```r
model.log <- lm(log.RT ~ AH4, data=d)
model.log2 <- lm(log.RT ~ AH4 + AH4.2, data=d)
summary(model.log2)
anova(model.log2, model.log)
```

```r
sdr <- rstudent(model.log)
threshold <- qt(1-(.05/2)/nrow(d),df=nrow(d)-2-1)
d[abs(sdr) >= threshold,]
d <- d[!(abs(sdr) >= threshold),]
```

```r
rs <- residuals(model.log)
hist(rs, prob=TRUE, breaks=100)
curve(dnorm(x,mean=0,sd=sd(rs)),add=TRUE)
```

```r
plot(model.log)
```

```r
shapiro.test(rs)
```

```r
null <- lm(RT ~ 1, data=d)
sse.null <- sum(residuals(null)^2)
sse.3 <- sum(rs^2)
ssr.y <- sse.null - sse.3
g <- rs^2 / (ssr.y / nrow(d))
hist(g)
d$g <- g
bp <- lm(g ~ AH4 + AH4.2, data=d)
sse.bp <- sum(residuals(bp)^2)
bp.null <- lm(g ~ 1)
summary(bp.null)
sse.bp.null <- sum(residuals(bp.null)^2)
ssr.g <- sse.bp.null - sse.bp
1-pchisq(ssr.g/2,df=2)
```
